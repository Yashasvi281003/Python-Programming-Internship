```python
import random

# Define word categories and clues
word_data = {
    "fruits": {
        "words": ["apple", "banana", "cherry", "date", "elderberry", "fig", "grape", "honeydew", "kiwi", "lemon"],
        "clue": "Category: Fruits"
    },
    "animals": {
        "words": ["elephant", "giraffe", "kangaroo", "penguin", "rhinoceros", "squirrel", "tiger", "zebra"],
        "clue": "Category: Animals"
    },
    "colors": {
        "words": ["blue", "green", "orange", "purple", "red", "yellow", "violet", "indigo", "pink"],
        "clue": "Category: Colors"
    }
}

def choose_category():
    print("Choose a word category:")
    for category in word_data:
        print(category)
    while True:
        category_choice = input().lower()
        if category_choice in word_data:
            return category_choice
        else:
            print("Invalid category. Please choose from the available categories.")

def choose_word(category):
    word_list = word_data[category]["words"]
    clue = word_data[category]["clue"]
    return random.choice(word_list), clue

def display_word(word, guessed_letters):
    display = ""
    for letter in word:
        if letter in guessed_letters:
            display += letter
        else:
            display += "_"
    return display

def hangman():
    category = choose_category()
    word_to_guess, clue = choose_word(category)
    guessed_letters = []
    attempts = 6
    score = 0

    print(f"Welcome to Hangman - Category: {category.capitalize()}")
    print(clue)

    while True:
        print("\nCurrent Score:", score)
        print("Guessed Letters:", guessed_letters)
        print("Attempts Left:", attempts)
        word_display = display_word(word_to_guess, guessed_letters)
        print("\nWord:", word_display)

        if "_" not in word_display:
            print("\nCongratulations! You guessed the word:", word_to_guess)
            score += 1
            play_again = input("Play again? (yes/no): ").lower()
            if play_again != "yes":
                break

        if attempts == 0:
            print("\nYou're out of attempts. The word was:", word_to_guess)
            play_again = input("Play again? (yes/no): ").lower()
            if play_again != "yes":
                break

        guess = input("Guess a letter: ").lower()

        if len(guess) != 1 or not guess.isalpha():
            print("Please enter a single letter.")
            continue

        if guess in guessed_letters:
            print("You already guessed that letter.")
            continue

        guessed_letters.append(guess)

        if guess in word_to_guess:
            print("Good guess!")
        else:
            attempts -= 1
            print("Incorrect guess.")

if __name__ == "__main__":
    hangman()
```
